# Import

```python
results = pd.read_csv("output/aggression_by_granularity_and_depth.csv")
```

```python
results.head(3)
```

# clean

```python
clean_results = results\
    .rename({'hand_strength':'Hand strength',
             'bet_probability':'P(BET)'}, axis = 1)
```

# Plots

```python
g = sns.FacetGrid(clean_results.loc[lambda x:x.depth == 2], col="granularity", col_wrap = 2)
g.map_dataframe(sns.barplot, x = "Hand strength", y = "P(BET)", hue = "context", palette = 'deep')
g.add_legend()
plt.savefig('output/aggression_by_granularity.png')
```

```python
g = sns.FacetGrid(clean_results.loc[lambda x:x.granularity == 5], col="depth", col_wrap = 2)
g.map_dataframe(sns.barplot, x = "Hand strength", y = "P(BET)", hue = "context", palette = 'deep')
g.add_legend()
plt.savefig('output/aggression_by_depth.png')
```
